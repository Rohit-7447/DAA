```python
def print_board(towers):
    for disk,name in towers.items():
        print("Disk:",disk," name:",name)
    print("-" * 25)
    
def towers_of_hanoi(n,source,auxillary,destination,towers):
    if n == 1:
        print("Move disk 1 from ",source," : ",destination)
        disk = towers[source].pop()
        towers[destination].append(disk)
        print_board(towers)
        return
        
    towers_of_hanoi(n-1,source,destination,auxillary,towers)  
    
    print("Move disk", n ,"from ",source," : ",destination)
    disk = towers[source].pop()
    towers[destination].append(disk)
    print_board(towers)
    towers_of_hanoi(n-1,auxillary,source,destination,towers)      
        

n = int(input("Enter the Number of disks:"))
towers = {
    'A' : list(range(n,0,-1)),
    'B' : [],
    'C' : []
}

print("Initial state:")
print_board(towers)

towers_of_hanoi(n,'A','B','C',towers)
```

Good afternoon, ma’am. This program demonstrates the Tower of Hanoi problem using recursion in Python. The Tower of Hanoi is a classic mathematical puzzle where we have three rods — usually named A, B, and C — and a certain number of disks stacked on one rod in decreasing size. The goal is to move all disks from the source rod to the destination rod, using the auxiliary rod as temporary storage. The rules are that only one disk can be moved at a time, and a larger disk can never be placed on top of a smaller one.

The program starts with the definition of the function print_board(towers). This function is used to display the current state of the three rods at any point in the game. It takes a dictionary named towers as input, where each key is a tower name like ‘A’, ‘B’, or ‘C’, and the value is a list containing the disks currently on that tower. Inside the function, a for loop goes through each tower and prints its name along with the disks on it. After displaying all three towers, I print a line of dashes to visually separate each step for clarity.

Next, we define the main function towers_of_hanoi(n, source, auxiliary, destination, towers). This is a recursive function that solves the problem step by step. It takes five parameters:

n, which is the number of disks to move,

source, the tower where the disks are initially placed,

auxiliary, the helper tower,

destination, the tower to which disks must be moved,

and towers, which keeps track of the current arrangement of all disks.

Inside this function, the first condition checks if n == 1. This is the base case of recursion — meaning only one disk is left to move. If this condition is true, the function prints a message saying “Move disk 1 from source to destination.” Then it removes the top disk from the source tower using pop() and adds it to the destination tower using append(). After the move, it calls print_board(towers) to display the new state of all three towers, and then returns to stop further recursion at this level.

If n is greater than 1, the function performs the recursive steps of the Tower of Hanoi logic.
First, it calls itself — towers_of_hanoi(n-1, source, destination, auxiliary, towers) — to move the top n-1 disks from the source tower to the auxiliary tower, temporarily using the destination tower. This step clears the way for the largest disk at the bottom to move.

After the recursive call finishes, it prints the instruction “Move disk n from source to destination,” which moves the largest disk from the source rod directly to the destination rod. This is done by popping the top disk from the source list and appending it to the destination list. After this move, it again calls print_board(towers) to show the updated positions of all disks.

Then comes the final recursive step — towers_of_hanoi(n-1, auxiliary, source, destination, towers) — which moves the smaller n-1 disks from the auxiliary tower to the destination tower, placing them neatly on top of the largest disk. In this way, the problem is broken down into smaller parts, and recursion continues until all disks are successfully transferred.

After defining the function, the program asks the user to input the number of disks using n = int(input("Enter the Number of disks:")). This allows the user to choose how many disks they want to solve the puzzle for. Then, it initializes the towers dictionary with three keys — ‘A’, ‘B’, and ‘C’.

Tower ‘A’ starts with all the disks arranged in descending order using list(range(n, 0, -1)),

Tower ‘B’ and ‘C’ are empty lists because initially, no disks are placed there.

Then, the program calls towers_of_hanoi(n, 'A', 'B', 'C', towers). This line starts the recursive process, moving all disks from tower A to tower C using tower B as a helper. During the process, every move is displayed step-by-step by the print_board() function.

After this main part, there’s an additional section for demonstration. The variable n_disks is set to 3, and an empty list moves_list is created. The program prints a message showing that it’s solving the Tower of Hanoi with 3 disks.
