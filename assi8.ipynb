```python
from collections import deque
def water_jug_bfs(jug1,jug2,goal):
    visited = set()
    queue = deque()
    queue.append((0,0))

    while queue:
        a , b = queue.popleft()
        if (a,b) in visited:
            continue
        visited.add((a,b)) 
        print(a, " | ", b)
        if a == goal or b == goal:
            print("\nReached the goal state!")
            return True
        next_states = set()    
        next_states.add((jug1, b))
        next_states.add((a, jug2))
        next_states.add((0, b))
        next_states.add((a, 0))
        pour = min(a, jug2 - b)
        next_states.add((a - pour, b + pour))
        pour = min(b, jug1 - a)
        next_states.add((a + pour, b - pour))

        for state in next_states:
            if state not in visited:
                queue.append(state)

    print("No solution possible.")
    return False

water_jug_bfs(4, 3, 2)
```

Good afternoon, ma’am. This program demonstrates the Water Jug problem using the Breadth-First Search (BFS) algorithm. The aim of this problem is to measure a specific amount of water using two jugs of different capacities. In this program, the BFS algorithm explores all possible water states step by step until it reaches the goal amount.

At the beginning, I import the deque class from the collections module. The deque is used as a queue that allows efficient insertion and deletion from both ends — which is perfect for BFS because BFS works in a first-in, first-out order.

Then, I define a function called water_jug_bfs(jug1, jug2, goal). This function takes three parameters — the capacity of jug1, the capacity of jug2, and the target goal amount of water we want to measure.

Inside the function, I first create an empty set called visited to keep track of all the states we have already explored. This helps prevent repeating the same states again and again. Then I create a queue using deque() and initialize it with the starting state (0, 0), which means both jugs are initially empty.

Now, I start a while loop that continues as long as there are states left in the queue. Inside the loop, I remove the first state from the queue using popleft() and store it in two variables, a and b, where a represents the amount of water in jug1 and b represents the amount of water in jug2.

Next, I check if this state (a, b) has already been visited. If it has, I skip it and continue to the next iteration. If it hasn’t been visited, I add it to the visited set so that we don’t visit it again in the future.

Then, I print the current state of the two jugs as a | b to show how much water each jug contains at that step. This allows us to visually track the process of filling and pouring water between the jugs.

Now comes the goal condition. If either jug contains exactly the goal amount of water — meaning a == goal or b == goal — then the program prints “Reached the goal state!” and returns True, indicating success.

If the goal hasn’t been reached yet, I create an empty set called next_states to store all possible new states that can be reached from the current state. Then I add all possible valid moves to this set.

The possible moves are as follows:

Fill Jug1 completely: (jug1, b)

Fill Jug2 completely: (a, jug2)

Empty Jug1: (0, b)

Empty Jug2: (a, 0)

Then I handle the two pouring actions:

Pour water from Jug1 into Jug2 until either Jug1 is empty or Jug2 is full. I calculate how much can be poured using pour = min(a, jug2 - b). Then I update the state to (a - pour, b + pour).

Pour water from Jug2 into Jug1 using a similar logic — pour = min(b, jug1 - a) — and the resulting state becomes (a + pour, b - pour).

After generating all these new states, I loop through them one by one. If a new state has not already been visited, I add it to the queue so it can be explored later. This ensures BFS explores all possible combinations of filling, emptying, and pouring water between the jugs.

If the queue becomes empty and the program hasn’t found any state where the goal amount is achieved, it prints “No solution possible.” and returns False.

Finally, outside the function, I call water_jug_bfs(4, 3, 2) which means the first jug has a capacity of 4 liters, the second jug has a capacity of 3 liters, and our goal is to measure exactly 2 liters of water.

When this program runs, it explores all possible combinations of filling and transferring water between the two jugs, printing each step until it reaches the state where one of the jugs contains 2 liters. Once it reaches that state, it prints that the goal has been reached and stops the search.

So, in summary, this program solves the classic Water Jug problem using the Breadth-First Search approach, which systematically explores all possible states level by level until it finds the shortest possible solution. It’s a perfect example of how BFS works in artificial intelligence by exploring all potential moves equally before going deeper, ensuring an optimal solution in the minimum number of steps.
